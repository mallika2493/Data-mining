```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
from scipy.stats import multivariate_normal
```

```python
####### Sampling on uniform distribution
```

```python
def uniform_sampling(x,y,num):
    res=[]
    for i in range(num):
        y_res = x+(y-x)*np.random.rand()
        
        res.append(y_res)
    return res
```

```python
import matplotlib.pyplot as plt
x_res = uniform_sampling(1,10,10000)
plt.hist(x_res)
plt.show()
```

```python
#Sampling with gaussian distribution 1 dimensional
```

```python
def guassian_distribution_sampling(mu,sigma,x):
    res = []
    y=multivariate_normal.pdf(x, mean=mu, cov=sigma)

    for i in range(len(y)):
        rand_y = np.random.rand()
        if rand_y < y[i]:
            res.append(x[i])
    return res
```

```python
res = guassian_distribution_sampling(3,1,x_res)
plt.hist(res)
plt.show()
```

```python
def gaussian_2D_distribution(mu,sigma,samples):
    res = []
    for i in range(samples):
        x_rand = np.random.uniform(0,10)
        y_rand = np.random.uniform(0,10)
        
        h = multivariate_normal.pdf([x_rand,y_rand], mean=mu, cov=sigma)
        y = np.random.uniform(0,1)
        
        if y<= h:
            
            res.append((x_rand,y_rand))
            plt.scatter(x_rand,y_rand)
    plt.show()
    return res
            
        
        
```
