```python
def get_random_mu(k,x):
    mu=[]
    for i in range(0, k):
        mu.append(x[np.random.randint(0, x.shape[0], size=1)])  
    #print(mu)
    return mu
```

```python
def generate_covariance(k):
    cov = []
    for i in range(k):
        cov.append(np.identity(2))
    return cov
```

```python
def EM(k,x,max_iterations):
    ###Initialization
    tol=0.001
    mu = get_random_mu(k,x)
    print("initial mu:",mu)
    if k == 2:
        cov = [np.identity(2),np.identity(2)]
    else:
        cov = [np.identity(2),np.identity(2),np.identity(2)]
    print("initial cov:",cov)
    N,d = x.shape
    w=[1./k]*k
    print("initial weights:",w)
    converged=False
    ll_old=0
    ll_new=0
    for iteration in range(max_iterations):
        
        ##The E step
        expectation = np.zeros((N, k))
        print("performing expectation for iteration:"+str(iteration))                       
        for i in range(0,N):
            for m in range(0,k):
                e_val=multivariate_normal.pdf(x[i], mean=mu[m][0], cov=cov[m])
                expectation[i][m]=w[m]*e_val
                 
            expectation[i] /=expectation[i].sum(0)
            
#         print (expectation)
        old_mu = mu
        old_cov = cov
        old_w = w
        
        print("old mu:",old_mu)
        print("old_cov:",old_cov)
        print("old_w:",old_w)
        
        mu = []
        cov = []
        w = []
        ##The M step
        print("performing maximization for iteration:"+str(iteration))                       
        for m in range(0,k):
            
            num_cov = 0.0
            num_mean = 0.0 
            expectation_sum = 0.0
            
            for i in range(0,N):
                
                temp = np.array(x[i]-old_mu[m])
                e=expectation[i][m]
                
                num_cov += e*np.matmul(temp.T,temp) 
                #print("intermediate cov:",num_cov)
                num_mean += e*x[i]
                expectation_sum+=e
            
            num_mean = [num_mean]
            cov.append(num_cov/expectation_sum)
            mu.append(num_mean/expectation_sum)
            w.append(expectation_sum/N)
            
        print("new mu:",mu)
        print("new_cov:",cov)
        print("old_w:",w)
        
        flag=False
        ll_new=0.0
        for i in range(N):
            for j in range(k):
                ll_new += math.log(expectation[i][j])
              
        if np.abs(ll_new-ll_old)<tol:
            print (ll_new,ll_old)
            break
            
        print("ll_old:"+str(ll_old))
        print("ll_new:"+str(ll_new)) 
        ll_old=ll_new     
         

        iteration+=1
    #print("ll_old:"+str(ll_old))
    #print("ll_new:"+str(ll_new))
    print("iteration:",iteration)
    print("results")
    for i in range(k):
        print("mu["+str(i)+"]:",mu[i])
        print("cov["+str(i)+"]:",cov[i])
        

        
        

            
            
       
    
    
    
```

```python
print("2 guassian")
start = time.time()
EM(2,x_2,50)
end = time.time()
print("Time taken:"+str(end - start)+" seconds")
```
